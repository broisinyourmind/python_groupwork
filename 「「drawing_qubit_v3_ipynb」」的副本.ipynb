```
def NewData(i):
    return np.array((I_std[i],Q_std[i])) 
def R(i):
    w = np.pi*2*0.023
    t = i
    c, s = np.cos(w*t), np.sin(w*t)
    return np.array(((c, -s), (s, c)))

r1_list,r2_list=[],[]
for i in range(T):
    tmp = np.dot(R(-i),NewData(i)) #做內積tmp:2*1 matrix
    r1_list.append(tmp[0])
    r2_list.append(tmp[1])
r1_list_mean=np.mean(r1_list)
r2_list_mean=np.mean(r2_list)
```

```
plt.xlim(0, 30)
plt.title('After rotation',fontdict={'size':20})
plt.plot(r1_list,label='cos')
plt.plot(r2_list,label='sin')
plt.legend()
plt.xlabel("t(ns)",fontdict={'size':12})
plt.ylabel("V",fontdict={'size':12})
```

```
#plt.title('Lissajous curve')
plt.plot(I_std,Q_std,label='before rotation')
plt.plot(r1_list,r2_list,label='after rotation')
plt.plot(r1_list_mean,r2_list_mean,'go', label='qubit state')
plt.legend()
#plt.title("")
plt.xlabel("cos",fontdict={'size':12})
plt.ylabel("sin",fontdict={'size':12})
```
